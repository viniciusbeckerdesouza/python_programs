```python
# Abra o arquivo HTML em modo de leitura
with open('page.html', 'r', encoding='utf-8') as arquivo:
    # Leia o conteúdo do arquivo
    conteudo_html = arquivo.read()

```

```python
from bs4 import BeautifulSoup

# Seu texto com marcações HTML
texto_html = """
<html>
<head>
    <title>Título da Página</title>
</head>
<body>
    <h1>Este é um exemplo</h1>
    <p>Este é um parágrafo com <a href="https://www.exemplo.com">um link</a>.</p>
</body>
</html>
"""

# Crie um objeto BeautifulSoup
soup = BeautifulSoup(conteudo_html, 'html.parser')

# Use o método get_text() para extrair apenas o texto sem as marcações HTML
texto_sem_marcações = soup.get_text()

# Imprima o texto sem as marcações HTML
print(texto_sem_marcações)
```

```python
texto_final = texto_sem_marcações.replace("        ", "\n")
texto_final = texto_sem_marcações.replace("    ", "\n")
print(texto_final)
```
